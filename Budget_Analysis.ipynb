```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("India_budget_2021.csv")
```

```python
data.head()
```

```python
print(data)
```

```python
# Dropping NULL Values
data = data.dropna()
```

```python
data = data.sort_values(by = 'Fund allotted(in ₹crores)', ascending = False)
```

```python
data
```

```python
# SET 1 
# Keeping data above 50000.00 255124

data1 = data[data['Fund allotted(in ₹crores)'] > 50000.00 ]
s = data[data['Fund allotted(in ₹crores)'] < 50000.00 ]['Fund allotted(in ₹crores)'].sum()
r = {'Department /Ministry' : 'OTHERS','Fund allotted(in ₹crores)': s}
data1 = pd.concat([data1, pd.DataFrame([r])], ignore_index = True)
data1
```

```python
data1 = data1.drop(index =0)
```

```python
data1
```

```python
# As Data is Ready - Let's try different plots to get a better visualisation of the budget distribution
```

```python
# Bar Plot
data1.plot.bar( x = 'Department /Ministry', y = 'Fund allotted(in ₹crores)' )
```

```python
# Donut Plot
df = data1["Fund allotted(in ₹crores)"]
labels = data1["Department /Ministry"]
plt.figure(figsize = (11,11))
plt.pie(df,labels = labels,autopct = '%1.1f%%', startangle = 80, pctdistance = 0.85, shadow = False)
central_circle = plt.Circle((0,0),0.5, color = 'white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size = 12)
plt.title("Distribution of Budget 2021", fontsize = 20)
plt.show()
```
